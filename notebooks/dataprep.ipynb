```python
# Load prepped_raw_data.xlsx into a dataframe. First row is header, second row should be skipped.
import numpy as np
import pandas as pd

import worldview
from worldview import preprocessor

# Show full dataframes when printing
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", lambda x: "%.5f" % x)
np.set_printoptions(threshold=100, suppress=True)

data_loc = worldview.get_data_dir()
df = pd.read_excel(data_loc / "prepped_raw_data.xlsx", header=0, skiprows=[1])

df = preprocessor.demographics_mapping(df)

df.head()
```

```python
for col in [
    "age_group",
    "gender",
    "transsexual",
    "ethnicity",
    "ethnicity_specify",
    "education",
    "education_other",
    "education_cleaned",
    "religious_spiritual_orientation",
    "religious_spiritual_orientation_other",
    "consider_open_inclusive",
    "experience_open_inclusive",
    "feel_experience_changed",
]:
    # count unique values in each column
    print(f"Column: {col}")
    print(df[col].value_counts(dropna=False))
    print("\n")
```

```python
# Add scale totals and dominant worldview to dataframe
df = preprocessor.calculate_worldview_scores(df)
df = preprocessor.qes_scores(df)
df = preprocessor.isi_subscale_scores(df)
df.head()
```

```python
df.to_csv(data_loc / "processed_data.csv")

keep_cols = [
    "age_group",
    "age_group_ordinal",
    "gender",
    "transsexual",
    "ethnicity",
    "education",
    "education_cleaned",
    "religious_spiritual_orientation",
    "feel_experience_changed",
    "consider_open_inclusive",
    "experience_open_inclusive",
    "traditional_wvs_total",
    "modern_wvs_total",
    "postmodern_wvs_total",
    "integrative_wvs_total",
    "dominant_worldview",
    "qes_total",
    "isi_diffuse_avoidant_total",
    "isi_informational_total",
    "isi_normative_total",
]

df[keep_cols].to_csv(data_loc / "processed_filtered_cols.csv")
```
